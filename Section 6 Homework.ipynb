```python
movies2 = pd.read_csv("P4-Section6-Homework-Dataset.csv", encoding = 'latin1')
```

```python
movies2
```

```python
movies3 = movies2[(movies2.Studio == 'Buena Vista Studios') | (movies2.Studio == 'Fox') | (movies2.Studio == 'Paramount Pictures') | (movies2.Studio == 'Sony') | (movies2.Studio == 'Universal') | (movies2.Studio == 'WB')]
```

```python
z =sns.boxplot(data=movies3, y='Gross % US', x='Genre')
plt.rcParams['figure.figsize'] = 50,30
sns.stripplot(x='Genre', y='Gross % US', data=movies3, jitter=True, size=15, linewidth=0, hue = 'Studio', alpha=0.7)
plt.legend(frameon=True, fontsize=40, shadow=True)
plt.ylabel("Gross % US", fontsize=50, color="DarkBlue",fontname="Times New Roman")
plt.xlabel("Budget", fontsize=50, color="DarkBlue",fontname="Times New Roman")
plt.xticks(fontsize=30)
plt.yticks(fontsize=40)
sns.set_style('darkgrid')
```
